```python
# Inisialisasi dataset
data = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
print(f"Data: {data}")

# Jumlah tinggi dalam dataset 
jumlah_tinggi = sum(data)
print(f"Jumlah tinggi: {jumlah_tinggi}")

# banyak data dalam dataset 
banyak_data = len(data)
print(f"Banyak data: {banyak_data}")

# Jumlah semua nilai dalam dataset 
mean_value = jumlah_tinggi / banyak_data
print(f"Mean: {mean_value}")
```

```python
import matplotlib.pyplot as plt

# Create a histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='purple')

# Add a vertical line at the mean
plt.axvline(x=mean_value, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean_value:.2f}')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Mean')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Inisialisasi dataset
data = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
print(f"Data: {data}")

# Mengurutkan data dari terkecil hingga terbesar
urutan = sorted(data)
print(f"Urutan data: {urutan}")

# Jumlah total data dalam dataset 
n = len(urutan)
print(f"Banyak data: {n}")

index_tengah = n//2 # double slash // is used for floor division

# Jika data berjumlah genap, median adalah rata-rata dari dua nilai tengah 
if n % 2 == 0:
    print(f"Data berjumlah Genap")
    median1 = urutan[index_tengah]
    median2 = urutan[index_tengah - 1]
    print(f"Nilai tengah: {median1} {median2}")
    median_value = (median1 + median2) / 2
    
# Jika data berjumlah ganjil, median adalah nilai tengah
else:
    print(f"Data berjumlah Ganjil")
    median_value = urutan[index_tengah]
    print(f"Nilai tengah: {median_value}")

print(f"Median: {median_value}")
```

```python
import matplotlib.pyplot as plt

# Create a histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='blue')

# Add a vertical line at the mean
plt.axvline(x=median_value, color='red', linestyle='dashed', linewidth=1, label=f'Median = {median_value:.2f}')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Median')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Import library Counter untuk menghitung modus
from collections import Counter

# Inisialisasi dataset
data = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
print(f"Data: {data}")

counter = Counter(data)
mode_data = dict(counter) # ubah ke tipe data dictionary
max_counter = max(list(counter.values())) # mencari data yang memiliki counter paling banyak
print(f"Max Muncul: {max_counter}")

mode_value = []
for k, v in mode_data.items():
    print(f"{k} muncul {v}x")
    if v == max(list(counter.values())):
        mode_value.append(k)
    
print(f"Modus: {mode_value}")
```

```python
import matplotlib.pyplot as plt

# Create a histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

# Add a vertical line at the mean
plt.axvline(x=mode_value[0], color='red', linestyle='dashed', linewidth=1, label=f'Modus 1 = {mode_value[0]:.2f}')
plt.axvline(x=mode_value[1], color='blue', linestyle='dashed', linewidth=1, label=f'Modus 2 = {mode_value[1]:.2f}')
plt.axvline(x=mode_value[2], color='aqua', linestyle='dashed', linewidth=1, label=f'Modus 3 = {mode_value[2]:.2f}')
plt.axvline(x=mode_value[3], color='grey', linestyle='dashed', linewidth=1, label=f'Modus 4 = {mode_value[3]:.2f}')
plt.axvline(x=mode_value[4], color='yellow', linestyle='dashed', linewidth=1, label=f'Modus 5 = {mode_value[4]:.2f}')
plt.axvline(x=mode_value[5], color='silver', linestyle='dashed', linewidth=1, label=f'Modus 6 = {mode_value[5]:.2f}')
plt.axvline(x=mode_value[6], color='olive', linestyle='dashed', linewidth=1, label=f'Modus 7 = {mode_value[6]:.2f}')
plt.axvline(x=mode_value[7], color='brown', linestyle='dashed', linewidth=1, label=f'Modus 8 = {mode_value[7]:.2f}')
plt.axvline(x=mode_value[8], color='orange', linestyle='dashed', linewidth=1, label=f'Modus 9 = {mode_value[8]:.2f}')
plt.axvline(x=mode_value[9], color='purple', linestyle='dashed', linewidth=1, label=f'Modus 10 = {mode_value[9]:.2f}')
plt.axvline(x=mode_value[10], color='pink', linestyle='dashed', linewidth=1, label=f'Modus 11 = {mode_value[10]:.2f}')


# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Modes')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
